## ü´ß Limpeza dos dados 
### 1. `rend_escolas_2022`:
1.1. Remo√ß√£o de coluna extra de √≠ndice.

1.2. Altera√ß√£o na nomeclatura das colunas.

1.3. Remo√ß√£o das primeiras e √∫ltimas linhas indesejadas.

1.4. Reseta√ß√£o de √≠ndices.

1.5. Retirada do ponto e retorno da parte anterior ao ponto no c√≥digo das escolas, para ficar no mesmo modelo os dois dataframes.

1.6. Sele√ß√£o das colunas necess√°rias.

1.7. Renomea√ß√£o das colunas.

### 2. `microdados_ed_basica_2023`:
2.1. Transforma√ß√£o da coluna de c√≥digos das escolas em string para se igualar ao outro dataframe. 

2.2. Selec√£o das colunas necess√°rias.

2.3. Renomea√ß√£o das colunas.

2.4. Substitui√ß√£o de d√≠gitos pelos dados referentes.

### 3. Verifica√ß√£o se h√° c√≥digos de escolas duplicadas

```python
rend_escolas_2020_df = rend_escolas_2020_df.drop(columns=['Unnamed: 0.1']) 
rend_escolas_2020_df.columns = rend_escolas_2020_df.iloc[7].values 
rend_escolas_2020_df = rend_escolas_2020_df.drop(rend_escolas_2020_df.index[:8]) 
rend_escolas_2020_df = rend_escolas_2020_df.drop(rend_escolas_2020_df.index[-2:])
rend_escolas_2020_df.reset_index(drop=True, inplace=True) 
rend_escolas_2020_df['CO_ENTIDADE'] = rend_escolas_2020_df['CO_ENTIDADE'].astype(str).str.partition('.')[0] 
rend_escolas_2020_df = rend_escolas_2020_df[['NO_MUNICIPIO', 'CO_ENTIDADE', 'TIPOLOCA', 'Dependad', 'tap_FUN', 'tap_MED', 'tre_FUN', 'tre_MED', 'tab_FUN', 'tab_MED']] 
rend_escolas_2020_df = rend_escolas_2020_df.rename(columns={'NO_MUNICIPIO':'municipio',
                                                    'CO_ENTIDADE':'codigo_escola',
                                                    'TIPOLOCA':'zona_de_localizacao', 
                                                    'Dependad':'dependencia_administrativa',
                                                    'tap_FUN':'total_de_aprovacao_ensino_fundamental',
                                                    'tap_MED':'total_de_aprovacao_ensino_medio',
                                                    'tre_FUN':'total_de_reprovacao_ensino_fundamental',
                                                    'tre_MED':'total_de_reprovacao_ensino_medio',
                                                    'tab_FUN':'taxa_de_abandono_ensino_fundamental',
                                                    'tab_MED':'taxa_de_abandono_ensino_medio'}) 
rend_escolas_2020_df
```

```python
rend_escolas_2022_df = rend_escolas_2022_df.drop(columns=['Unnamed: 0.1']) 
rend_escolas_2022_df.columns = rend_escolas_2022_df.iloc[7].values 
rend_escolas_2022_df = rend_escolas_2022_df.drop(rend_escolas_2022_df.index[:8]) 
rend_escolas_2022_df = rend_escolas_2022_df.drop(rend_escolas_2022_df.index[-2:])
rend_escolas_2022_df.reset_index(drop=True, inplace=True) 
rend_escolas_2022_df['CO_ENTIDADE'] = rend_escolas_2022_df['CO_ENTIDADE'].astype(str).str.partition('.')[0] 
rend_escolas_2022_df = rend_escolas_2022_df[['NO_MUNICIPIO', 'CO_ENTIDADE', 'NO_CATEGORIA', 'NO_DEPENDENCIA', '1_CAT_FUN', '1_CAT_MED', '2_CAT_FUN', '2_CAT_MED', '3_CAT_FUN', '3_CAT_MED']] 
rend_escolas_2022_df = rend_escolas_2022_df.rename(columns={'NO_MUNICIPIO':'municipio',
                                                    'CO_ENTIDADE':'codigo_escola',
                                                    'NO_CATEGORIA':'zona_de_localizacao', 
                                                    'NO_DEPENDENCIA':'dependencia_administrativa',
                                                    '1_CAT_FUN':'total_de_aprovacao_ensino_fundamental',
                                                    '1_CAT_MED':'total_de_aprovacao_ensino_medio',
                                                    '2_CAT_FUN':'total_de_reprovacao_ensino_fundamental',
                                                    '2_CAT_MED':'total_de_reprovacao_ensino_medio',
                                                    '3_CAT_FUN':'taxa_de_abandono_ensino_fundamental',
                                                    '3_CAT_MED':'taxa_de_abandono_ensino_medio'}) 
rend_escolas_2022_df
```
